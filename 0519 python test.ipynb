```python
import matplotlib.pyplot as plt
#import seaborn as sns
import pandas as pd
tdata = pd.read_csv("cases_add_Taiwan.csv", encoding = 'big5')
tdata[['Date', 'Worldwide']].plot(kind = 'line', x = 'Date', y = 'Worldwide', title = "COVID-19 Daily Report")
plt.show()
```

```python
plt.plot(tdata["Date"], tdata["Taiwan"],c = "r",label = "Taiwna")
plt.plot(tdata["Date"], tdata["Australia"],c = "g")
plt.plot(tdata["Date"], tdata["Germany"],c = "b")
plt.xlabel("Date", fontweight = "bold")
plt.ylabel("Count", fontweight = "bold")
plt.xticks(rotation=45)
plt.show()
```
